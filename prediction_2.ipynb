Predicting Winner of Chinese GP 2025 - second round 

```python
import fastf1 
import pandas as pd 
import numpy as np 
```

```python
session_2024 = fastf1.get_session(2024, 'Chinese Grand Prix', 'R')
session_2024.load()
```

```python
session_2024.laps[:5].columns 
```

```python
session_2024.laps.head(4)
```

```python
laps_2024 = session_2024.laps[['Driver' ,'LapTime','Sector1Time', 'Sector2Time', 'Sector3Time']].copy()
laps_2024
```

```python
columns = ['LapTime', 'Sector1Time', 'Sector2Time', 'Sector3Time']
for col in columns:
    laps_2024[f"{col}  (s)"] = laps_2024[col].dt.total_seconds()
```

```python
laps_2024
```

```python
laps_2024.dropna(inplace=True)
```

```python
laps_2024.columns
```

```python
sector_2024 = laps_2024.groupby('Driver')[['LapTime  (s)', 'Sector1Time  (s)', 'Sector2Time  (s)', 'Sector3Time  (s)']].mean().sort_values(by='LapTime  (s)', ascending=True).head(10)
```

```python
sector_2024['LapTime  (s)']
```

Getting qualified data 

```python
session_2025_q = fastf1.get_session(2025, 'Chinese Grand Prix', 'Q')
session_2025_q.load()
```

```python
results_2025_q = session_2025_q.results[['Abbreviation','Q3']].copy()
results_2025_q['Q3'] = results_2025_q['Q3'].dt.total_seconds()
results_2025_q.dropna(inplace=True)
```

```python
results_2025_q.rename(columns={'Abbreviation': 'Driver', 'Q3': 'Q3  (s)'}, inplace=True)
```

```python
results_2025_q
```

```python
merged_data = results_2025_q.merge(sector_2024, on='Driver', how='left')
merged_data.columns
```

```python
X = merged_data[["Q3  (s)", "Sector1Time  (s)", "Sector2Time  (s)", "Sector3Time  (s)"]].fillna(0)
X
```

```python
y = sector_2024.groupby("Driver")["LapTime  (s)"].mean().reset_index()["LapTime  (s)"]
y.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# Train Gradient Boosting Model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=38)
model = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, random_state=38)
model.fit(X_train, y_train)
```

```python
predicted_race_time = model.predict(X)
predicted_race_time
```

```python
results_2025_q["PredictedRaceTime (s)"] = predicted_race_time
results_2025_q.sort_values(by="PredictedRaceTime (s)", ascending=True, inplace=True)
```

```python
results_2025_q
```

```python
print("\n🏁 Predicted 2025 Chinese GP Winner with New Drivers and Sector Times 🏁\n")
print(results_2025_q[["Driver", "PredictedRaceTime (s)"]])
```

```python
from sklearn.metrics import mean_absolute_error
# Evaluate Model
y_pred = model.predict(X_test)
print(f"\n🔍 Model Error (MAE): {mean_absolute_error(y_test, y_pred):.2f} seconds")
```
